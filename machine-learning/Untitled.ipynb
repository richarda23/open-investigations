```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.arange(0,10*math.pi,0.1)
```

```python
def update_w_and_b(xs, values, w, b, alpha):
    dl_dw = 0.0
    dl_db = 0.0
    N = len(xs)
    for i in range(N):
        dl_dw += -2 *xs[i] *(values[i] - (w * xs[i] + b))
        dl_db += -2  *(values[i] - (w * xs[i] + b))
    w = w -  (1 / float(N)) * dl_dw * alpha
    b = b- (1 / float(N)) * dl_db * alpha
    return w,b
```

```python
def train (xs, values, w,b, alpha, epochs):
    plt.scatter(xs, values)
    for e in range(epochs):
        w,b = update_w_and_b(xs, values, w, b, alpha)
        if e % 1000 == 0:
            print (f"Epoch {e}, loss = {avg_loss(xs, values, w, b)}")
            predicted = xs * w +  b
            plt.plot(xs, predicted, label=str(e))
    return w,b
```

```python
def avg_loss(xs, values, w, b):
    N = len(xs)
    totalError = 0.0
    for i in range(N):
        y = (w * i +b)
        totalError += (values[i] - y) **2
    return totalError / float(N)

def predict (x,w,b):
    return w*x + b
```

```python
deltas = (np.random.random(15) * 8) -3
values = np.linspace(8,50,15) + deltas
xs = np.arange(15)
print (values)
print (xs)
```

```python
w,b=train(xs, values, 0.0, 0.0, 0.001, 20000)
print(w,b)
```

```python
predict(13,w ,b)
```
